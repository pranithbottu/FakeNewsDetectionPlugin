```python
import pandas as pd
import os

testing_3a = pd.read_csv("./Task3a_testing.csv")
testing_3b = pd.read_csv("./Task3b_testing.csv")
training_3a = pd.read_csv("./Task3a_training.csv")
training_3b = pd.read_csv("./Task3b_training.csv")
```

```python
print(training_3a.head())
count_all = {'"': 0, '[END]': 0}
rating_count = {
    "false": {'"': 0, '[END]': 0}, 
    "true": {'"': 0, '[END]': 0}, 
    "partially false": {'"': 0, '[END]': 0}, 
    "other": {'"': 0, '[END]': 0}, 
}
for i in range(len(training_3a["text"])):
    rating = training_3a["our rating"][i].lower()
    if isinstance(training_3a["title"][i], float):
        training_3a["title"][i] = ""
    for w in training_3a["title"][i].split(" ") + training_3a["text"][i].split(" "):
        w = w.lower()
        if len(w) == 0:
            continue
        if w[0] in ['"', "'"] or w[-1] in ['"', "'"]:
            count_all['"'] += 1
            rating_count[rating]['"'] += 1
            w.replace('"', '').replace("'", '')
        if w[-1] in ['.', "!", "?"]:
            count_all['[END]'] += 1
            rating_count[rating]['[END]'] += 1
            w = w[:-1]
        
        w = ''.join(filter(lambda i: str.isalnum(i) or i == "'", w))
        if w in count_all:
            count_all[w] += 1
        else:
            count_all[w] = 1
        if w in rating_count[rating]:
            rating_count[rating][w] += 1
        else:
            rating_count[rating][w] = 1
```

```python
print(len(count_all))
s = []
for w in count_all:
    if count_all[w] < 5:
        s.append(w)
for w in s:
    print(w)
    del count_all[w]
print(len(count_all))
```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer
from nltk import word_tokenize, pos_tag
from collections import defaultdict
tag_map = defaultdict(lambda : wn.NOUN)
tag_map['J'] = wn.ADJ
tag_map['V'] = wn.VERB
tag_map['R'] = wn.ADV

def update_counts(text, dict_updates):
    tokens = word_tokenize(text.lower())
    lemma_function = WordNetLemmatizer()
    for token, tag in pos_tag(tokens):
        if len(token) > 1:
            token = token.replace('.', '')
            token = token.replace('-', ' ').strip()
            if token == "":
                continue
            if token[0] in ["'", '"']:
                token = token[1:]
            if token[-1] in ["'", '"']:
                token = token[:-1]
            
        lemma = lemma_function.lemmatize(token, tag_map[tag[0]])
        for d in dict_updates:
            if lemma in d:
                d[lemma] += 1
            else:
                d[lemma] = 1
    return

d = {}
rating_count = {
    "false": {}, 
    "true": {}, 
    "partially false": {}, 
    "other": {}, 
}
for i in range(len(training_3a["text"])):
    rating = training_3a["our rating"][i].lower()
    if isinstance(training_3a["title"][i], float):
        training_3a["title"][i] = "NOTITLE"
    text = training_3a["title"][i] + " " + training_3a["text"][i]
    updates = [d, rating_count[rating]]
    update_counts(text, updates)
```

```python
s = []
print(len(d))
for w in d:
    if d[w] < 5:
        s.append(w)
for w in s:
    print(w)
    del d[w]
print(len(d))
```
